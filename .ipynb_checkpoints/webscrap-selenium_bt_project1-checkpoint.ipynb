```python
# üìö Basic libraries
import pandas as pd
import os
import random
import re
from bs4 import BeautifulSoup

#‚ùóNew Libraries !
import time
from getpass import getpass # to safely storage your password
# selenium libraries
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
import pathlib
```

```python
# ‚öôÔ∏è Settings
pd.set_option('display.max_columns', None) # display all columns
import warnings
warnings.filterwarnings('ignore') # ignore warnings
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
driver.get('https://www.bundestag.de/abgeordnete')
```

```python
time.sleep(random.random() * 3)
```

```python
mdb_dict = {}

# find button "Listenansicht"
liste_button = driver.find_element(By.XPATH, '//a[@title="Listenansicht"]')
    
# click button
liste_button.click()

# time to load website
time.sleep(2)

# find the names of the MdBs
abgeordnete = driver.find_elements(By.XPATH, '//div[@class="bt-teaser-person-text"]/h3')

# find the parties of the MdBs
parteien = driver.find_elements(By.XPATH, '//p[@class="bt-person-fraktion"]')

# getting the names and parties, making sure len is the same and adding them to dict
if len(abgeordnete) == len(parteien):
    for name, partei in zip(abgeordnete, parteien):
        if name.text.strip() and partei.text.strip():  # checking they are not empty
            mdb_dict[name.text] = partei.text # adding to dict

# Schlie√üe den Browser
driver.quit()

# Die gesammelten Namen anzeigen
print(mdb_dict)
```

```python
# convert dict to Dataframe

df_namen = pd.DataFrame(list(mdb_dict.items()), columns=['Name', 'Partei'])
```
