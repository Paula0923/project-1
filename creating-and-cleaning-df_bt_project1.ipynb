## merging the new df ("df_names") with df_main
- all values of df_names should replace those in the corresponding columns in df_main ("Name" and "MdB"), the rest of the info should stay the same
- also, all values in "Names" which are in df_names but not in df_main should be added as rows to the df
- in the cleaning afterwards all rows without the value "MdB" in the column "MdB" should be erased to have only the "real" MdBs in the df; to achieve this columns "MdB_x" and "Partei_x" will be deleted because this info is from df_main and not updated

```python
# first create a copy of both dfs to do the combining on
df_main2 = df_main.copy()
df_names2 = df_names.copy()

# Setting the index of both dfs to "Name"
df_main2.set_index("Name", inplace=True)
df_names2.set_index("Name", inplace=True)

# merges the dataframes, df_names should dominate, therefore "right" merge
df_main_new = pd.merge(df_main,df_names, on="Name", how="right")

# reset index
df_main_new.reset_index(inplace=True)

df_main_new.head()
```

### Extracting academical grade from "Vorname"

```python
# Extracting grade from "Vorname" and adding to new column "Akadem. Grad"
df_main_new['Akadem. Grad'] = df_main_new['Vorname'].str.extract(r'^(Dr\.|Prof\.)')

# Erasing academical grade from "Vorname"
df_main_new['Vorname'] = df_main_new['Vorname'].str.replace(r'^(Dr\.|Prof\.)\s*', '', regex=True)

df_main_new
```
